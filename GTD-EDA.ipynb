```python
gtd.drop(['imonth', 'iday', 'approxdate'], axis=1, inplace=True)
#while time might be an interesting avenue for which to estimate, with a number of nulls and difficulties
#in the way to impute them, I feel it better just to drop them entirely. Maybe come back when I have more time?
```

```python
gtd['extended'].unique()
#binary variable - might be important for the type of attack, which is relevant to us, and has no nulls,
#so we'll keep this variable around
```

```python
gtd.drop('resolution', axis=1, inplace=True)
#this variable is a text variable that is only even relevant when extended is true, so we drop it
#next are country and region are obviously important - text less so in making any sort of inference,
#but helpful in double checking any conclusions later to make sure that they're reasonable
```

```python
#next are provstate and city, which are both text fields that can't be used for inference
gtd.drop(['provstate', 'city'], axis=1, inplace=True)
```

```python
#latitude and longitude are numerical values, but can we impute them using the means of the latitudes and longitudes
#of whichever country the event occurred within, even if this might not be the most accurate measure - it could
#still be a reasonable measure to take.
for country in gtd['country'].unique():
    if len(gtd[gtd['country'] == country]) == gtd[gtd['country'] == country]['latitude'].isnull().sum():
        print(country)
```

```python
#this is the only country where we don't have any latitude and longitudes for, but Andorra is small and can be
#input manually without too much hassle
gtd[gtd['country'] == 7]
```

```python
for i in gtd[gtd['latitude'].isnull() & gtd['longitude'].isnull()].index:
    country = gtd.loc[i]['country']
    mean_lat = gtd[gtd['country'] == country]['latitude'].mean()
    mean_long = gtd[gtd['country'] == country]['longitude'].mean()
    gtd.loc[i, 'latitude'] = mean_lat
    gtd.loc[i, 'longitude'] = mean_long
```

```python
gtd['latitude'].isnull().sum()
```

```python
#did that get everything aside from Andorra?
gtd[gtd['latitude'].isnull()]
```

```python
#It did! Now to manually input the last value based off of the coordinates of Andorra's capital city
gtd.loc[2397, 'latitude'] = 42 + (30/60)
gtd.loc[2397, 'longitude'] = 1 + (31/60)
```

```python
gtd.loc[2397]
```

```python
#in regards to specificity, 5 is the value for locations that are unknown, so we'll use that one
gtd['specificity'] = gtd['specificity'].fillna(5)
gtd['specificity'] = gtd['specificity'].astype(int)
```

```python
for col in gtd.columns:
    print(col + ": " + str(gtd[col].isnull().sum()) + " nulls")
```

```python
#location is a text entry that doesn't let us infer anything
gtd.drop('location', axis=1, inplace=True)
```

```python
#so is summary
gtd.drop('summary', axis=1, inplace=True)
```

```python
#alternative and alternative text is only relevant in cases where terrorism is doubted - can be dropped
gtd.drop(['alternative', 'alternative_txt'], axis=1, inplace=True)
```

```python
#not all attacks had multiple avenues, so we can dump everything related to attacks past 1, but we can record whether
#an attack had multiple avenues
multiatk = []
for i in gtd['attacktype2'].isnull():
    if i:
        multiatk.append(0)
    else:
        multiatk.append(1)
pd.Series(multiatk).value_counts()
```

```python
#Everything else related to multiple attack types needs to go
gtd['attacktype2'] = multiatk
gtd.rename(columns={'attacktype2':'multiatk'}, inplace=True)
gtd.drop(['attacktype2_txt', 'attacktype3', 'attacktype3_txt'], axis=1, inplace=True)
```

```python
#not all target types have subtypes, making that impossible to impute
gtd.drop(['targsubtype1', 'targsubtype1_txt'], axis=1, inplace=True)
```

```python
#corp1 and target1 are text variables
gtd.drop(['corp1', 'target1'], axis=1, inplace=True)
```

```python
#for simplicity's sake, we'll assume that any unknown target nationality was that of where the attack occurred
for i in gtd[gtd['natlty1'].isnull()].index:
    natlty = gtd.loc[i, 'country']
    natlty_txt = gtd.loc[i, 'country_txt']
    gtd.loc[i, 'natlty1'] = natlty
    gtd.loc[i, 'natlty1_txt'] = natlty_txt
gtd['natlty1'] = gtd['natlty1'].astype(int)
```

```python
#same thing we did for attack types we can do for targets - check whether there was more than one
multitarg = []
for i in gtd['targtype2'].isnull():
    if i:
        multitarg.append(0)
    else:
        multitarg.append(1)
pd.Series(multitarg).value_counts()
```

```python
#we're using the same process as we did before
gtd['targtype2'] = multitarg
gtd.rename(columns={'targtype2':'multitarg'}, inplace=True)
gtd.drop(['targtype2_txt', 'targsubtype2', 'targsubtype2_txt', 'corp2', 'target2', 'natlty2', 'natlty2_txt'], axis=1, inplace=True)
gtd.drop(['targtype3', 'targtype3_txt', 'targsubtype3', 'targsubtype3_txt', 'corp3', 'target3', 'natlty3', 'natlty3_txt'], axis=1, inplace=True)
```

```python
for col in gtd.columns:
    print(col + ": " + str(gtd[col].isnull().sum()) + " nulls")
```

```python
#we're gonna do the same thing to group that we did to the others earlier
multigrp = []
for i in gtd['gname2'].isnull():
    if i:
        multigrp.append(0)
    else:
        multigrp.append(1)
pd.Series(multigrp).value_counts()
```

```python
gtd['gsubname'] = multigrp
gtd.rename(columns={'gsubname':'multigrp'}, inplace=True)
gtd.drop(['gname2', 'gsubname2', 'gname3', 'gsubname3'], axis=1, inplace = True)
```

```python
#motive can be dropped without further consequence, text variable
gtd.drop('motive', axis=1, inplace = True)
```

```python
#presuming that if the value is null, that they don't actually know whether they were confirmed or not, 
#so we'll set guncertain1 to be 1 (designates that there is uncertainty)
gtd['guncertain1'] = gtd['guncertain1'].fillna(1)
gtd['guncertain1'] = gtd['guncertain1'].astype(int)
```

```python
#don't need these
gtd.drop(['guncertain2', 'guncertain3'], axis=1, inplace=True)
```

```python
#these have inbuilt unknowns, but as a numerical instead of categorical variable, that might throw things off
#so we'll drop
gtd.drop(['nperps', 'nperpcap'], axis=1, inplace=True)
```

```python
#impossible to reasonably impute missing claims, so all get dropped
gtd.drop(['claimed', 'claimmode', 'claimmode_txt', 'claim2', 'claimmode2', 'claimmode2_txt', 'claim3', 'claimmode3', 'claimmode3_txt', 'compclaim'], axis=1, inplace=True)
```

```python
for col in gtd.columns:
    print(col + ": " + str(gtd[col].isnull().sum()) + " nulls")
```

```python
#I think you know what we're going to do with weapons
multiweap = []
for i in gtd['weaptype2'].isnull():
    if i:
        multiweap.append(0)
    else:
        multiweap.append(1)
pd.Series(multiweap).value_counts()
```

```python
#subtypes and details can't be filled, so they get dropped too
gtd['weaptype2'] = multiweap
gtd.rename(columns={'weaptype2':'multiweap'}, inplace=True)
gtd.drop(['weapsubtype1', 'weapsubtype1_txt', 'weapsubtype2', 'weapsubtype2_txt', 'weapsubtype3', 'weapsubtype3_txt', 'weapsubtype4', 'weapsubtype4_txt'], axis=1, inplace=True)
gtd.drop(['weaptype2_txt', 'weaptype3', 'weaptype3_txt', 'weaptype4', 'weaptype4_txt'], axis=1, inplace=True)
gtd.drop('weapdetail', axis=1, inplace=True)
```

```python
gtd['property'].value_counts()
```

```python
#I don't want to get rid of nkill, as it's an important subject. Though it may be inaccurate, we'll presume
#for the purposes that an unknown number of casualties is 0.
gtd['nkill'] = gtd['nkill'].fillna(0)
gtd['nkill'] = gtd['nkill'].astype(int)
```

```python
#same for nwound
gtd['nwound'] = gtd['nwound'].fillna(0)
gtd['nwound'] = gtd['nwound'].astype(int)
```

```python
gtd.drop(['nkillus', 'nkillter'], axis=1, inplace=True)
gtd.drop(['nwoundus', 'nwoundte'], axis=1, inplace=True)
```

```python
#since this is categorical, we can set the unknowns to the variable that is unknown
gtd['propextent'].value_counts()
```

```python
gtd['propextent_txt'].value_counts()
```

```python
#these had 0 damage, so we're putting them in the lowest category
for i in gtd[gtd['property'] == 0].index:
    gtd.loc[i, 'propextent'] = 3
    gtd.loc[i, 'propextent_txt'] = "Minor (likely < $1 million)"
```

```python
for i in gtd[gtd['property'] == -9].index:
    gtd.loc[i, 'propextent'] = 4
    gtd.loc[i, 'propextent_txt'] = "Unknown"
```

```python
for i in gtd[gtd['propextent'].isnull()].index:
    gtd.loc[i, 'propextent'] = 4
    gtd.loc[i, 'propextent_txt'] = "Unknown"
```

```python
gtd['propextent'] = gtd['propextent'].astype(int)
```

```python
for col in gtd.columns:
    print(col + ": " + str(gtd[col].isnull().sum()) + " nulls")
```

```python
#propvalue is difficult to impute, propcomment is a text variable unable to be predictive
gtd.drop(['propvalue', 'propcomment'], axis=1, inplace=True)
```

```python
#lots of these following variables are only applicable for specific types of terrorist events, so we can drop them
gtd.drop(['ishostkid', 'nhostkid', 'nhostkidus', 'nhours', 'ndays', 'divert', 'kidhijcountry'], axis=1, inplace=True)
gtd.drop(['ransom', 'ransomamt', 'ransomamtus', 'ransompaid', 'ransompaidus', 'ransomnote', 'hostkidoutcome', 'hostkidoutcome_txt', 'nreleased'], axis=1, inplace=True)
```

```python
#we can drop the notes, they're text values that don't hold predictive value
gtd.drop(['addnotes', 'scite1', 'scite2', 'scite3'], axis=1, inplace=True)
```

```python
#dbsource just tells us which source we're getting it from and has no predictive value
gtd.drop('dbsource', axis=1, inplace=True)
```

```python
#while knowing if there are related events is important, we don't need the information inside the variable. 
#So we'll change this to a 1/0 binary variable based on isnull(), like we've done before.
related = []
for i in gtd['related'].isnull():
    if i:
        related.append(0)
    else:
        related.append(1)
gtd['related'] = related
gtd['related'] = gtd['related'].astype(int)
```
